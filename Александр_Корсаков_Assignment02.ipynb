### 1. Одномерная линейная регрессия.
Вывести формулу линейной регрессии для одномерного случая. Дана

### 2. Оптимальный шаг градиентного спуска. 
Рассмотрим задачу обучения линейной регрессии:

$$
Q(\vec{w})=(X\vec{w}-\vec{y})^T(X\vec{w}-\vec{y}) \to \min_{w},
$$
и будем решать ее с помощью градиентного спуска. Допустим, мы находимся на некоторой итерации $k$, и хотим выполнить очередной шаг:
$$
\vec{w}^{(k)}=\vec{w}^{(k-1)}-\eta \nabla_w Q(\vec{w}^{(k-1)}).
$$
При известных $ \vec{y}, X,\vec{w}^{(k-1)}$, определите длину шага $\eta$, при которой уменьшение значение функционала будет наибольшим, т.е.: 

$$
Q(\vec{w}^{(k-1)}-\eta \nabla_w Q(\vec{w}^{(k-1)})) \to \min_{\eta}.
$$

## Одномерная линейная регрессия

$$
  \begin{equation*}
  X = \left(
  \begin{array}{cccc}
  1 & x_1 \\
  1 & x_2\\
  \vdots & \vdots\\
  1 & x_N
  \end{array}
  \right)
  \;\;\;;
  w = \left(
  \begin{array}{cccc}
  w_0 & w_1
  \end{array}
  \right)
  ;\;\;\;
  Y = \left(
  \begin{array}{cccc}
  y_1 \\
  y_2\\
  \vdots\\
  y_N
  \end{array}
  \right)
  \end{equation*}
$$


$$
  Xw=Y
$$

$Q(w)=\sum^{N}_{i=1} (x_iw^*-y_i)^2 \rightarrow min$

\\
Для одномерного случая

\\
$Q(w_0^*,w_1^*)=\sum^{N}_{i=1} (x_iw_1^*+w_0^*-y_i)^2 \rightarrow min$

\\
Запишем необходимые условия экстремума по $w_0^*, \; w_1^*$

\\
$$
  \begin{equation*}
  \left\{
  \begin{array}{lr}
  \frac{\partial Q}{\partial w_0^*} = -2*\sum_{i=1}^N(y_i-w_0^*-w_1^*x_i)=0 \\
  \frac{\partial Q}{\partial w_1^*} = -2*\sum_{i=1}^Nx_i(y_i-w_0^*-w_1^*x_i)=0
  \end{array}
  \right.
  \end{equation*}
$$

\\
Раскрывая скобки получаем

\\


$$
  \begin{equation*}
  \left\{
  \begin{array}{lr}
  Nw_0^*+w_1^*\sum_{i=1}^Nx_i=\sum_{i=1}^Ny_i\\
   (\sum_{i=1}^Nx_i)w_0^*+w_1^*\sum_{i=1}^Nx_i^2=\sum_{i=1}^Nx_iy_i\\
  \end{array}
  \right.
  \end{equation*}
$$

\\

$Из \; первого \; уравнения \; выражаем\; w_0^*, \; из \; второго \; w_1^*$

\\

$$
  w^*_{1.мнк}=\frac{N\sum_{i=1}^Nx_iy_i-(\sum_{i=1}^Nx_i)(\sum_{i=1}^Ny_i)}{N\sum_{i=1}^Nx_i^2-(\sum_{i=1}^Nx_i)^2}=\frac{\sum_{i=1}^N(x_i-\overline{x})(y_i-\overline{y})}{\sum_{i=1}^N(x_i-\overline{x})^2}
$$

\\
$$
  w^*_{0.мнк}=\overline{y}-w^*_{1.мнк}\overline{x}
$$

\\
Кратко для общей формы
$$
  Q(\widehat{W})=Y^TY-2\widehat{W}^TX^TY+\widehat{W}^TX^TX\widehat{W}
$$

\\
$$
  \frac{\partial Q(\widehat{W})}{\partial \widehat{W}} = (\frac{\partial Q(\widehat{W})}{\partial \widehat{w_0}},\frac{\partial Q(\widehat{W})}{\partial \widehat{w_1}},\dots,\frac{\partial Q(\widehat{W})}{\partial \widehat{w_p}})^T
$$

\\
$$
  \frac{\partial Q(\widehat{W})}{\partial \widehat{W}} =-2X^TY+2X^TX\widehat{W}=0
$$

\\
$$
  X^TX\widehat{W}=X^TY
$$

\\
$$
  \widehat{W}=(X^TX)^{-1}X^TY
$$

## Оптимальный шаг градиентного спуска

\\
Рассмотрим функцию одной переменной
$$
g(\eta)=Q(\vec{w}^{(k-1)}-\eta \nabla_w Q(\vec{w}^{(k-1)})) \to \min_{\eta}.
$$

\\
Так как имеем дело с квадратичным функционалаом, то ряд тейлора для введённой функции примет вид

\\
$$
  g(\eta)=Q(\vec{w}^{(k-1)}-\eta \nabla_w Q(\vec{w}^{(k-1)}))=Q(\vec{w}^{(k-1)})+(-\eta) \nabla_w Q(\vec{w}^{(k-1)})\nabla_w Q(\vec{w}^{(k-1)})+1/2*(\eta)^2*(\nabla_w Q(\vec{w}^{(k-1)}))^T (\nabla^2_w Q(\vec{w}^{(k-1)})) \nabla_w Q(\vec{w}^{(k-1)})
$$

\\
Берём производную по длине шага и приравниваем нулю, для выполнения необходимого условия экстремума

\\
$$
  g^{,}(\eta)=-\nabla_w Q(\vec{w}^{(k-1)})*\nabla_w Q(\vec{w}^{(k-1)})+\eta*(\nabla_w Q(\vec{w}^{(k-1)}))^T (\nabla^2_w Q(\vec{w}^{(k-1)}))( \nabla_w Q(\vec{w}^{(k-1)})=0
$$

\\
Тогда получаем, что оптимальная длина к-ого шага
$$
  \eta = \frac{\nabla_w Q(\vec{w}^{(k-1)})*\nabla_w Q(\vec{w}^{(k-1)})}{(\nabla_w Q(\vec{w}^{(k-1)}))^T (\nabla^2_w Q(\vec{w}^{(k-1)})) (\nabla_w Q(\vec{w}^{(k-1)})}
$$

В большинстве случаев данный метод представляется затруднительным для применения, потому что необходимо быть в состоянии посчитать гессиан

### 3. Стохастический градиентный спуск 
Напишите функцию стохастического градиентного спуска. 
Функция использует для оценки градиента всего функционала ошибки значение градиента, вычисленного по подвыборке размера batch_size=10. Визуализируйте траекторию, длина шага $\eta=1$.

```python
np.random.seed(17)
n_features = 2
n_objects = 300
w_true = np.array([1.0,1.0])
X = np.random.uniform(-5,5, (n_objects, n_features))
Y = X.dot(w_true) + np.random.normal(0,1, (n_objects))
```

```python
#ваш код здесь
def stochastic(X, Y, w_0, step_size, num_steps):
  w = w_0.copy()
  batch_size=10
  w_list = []
  num = np.random.randint(0, X.shape[0]-1, size = (num_steps,batch_size))
  for i in num:
    w_list.append(w)
    w = w - 2*step_size*(np.dot(X[i].T, np.dot(X[i],w)-Y[i]))/batch_size
  return w_list
```

```python
num_steps = 20
step_size = 1e-2
w_0 = np.array([0,0])


w_list=stochastic(X,Y,w_0,step_size, num_steps)
w_list = np.array(w_list)
print(w_list)

A, B = np.meshgrid(np.linspace(0, 2, 100), np.linspace(0, 2, 100))

levels = np.empty_like(A)
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        w_tmp = np.array([A[i, j], B[i, j]])
        levels[i, j] = np.mean(np.power(np.dot(X, w_tmp) - Y, 2))


plt.figure(figsize=(12, 8))
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.gca().set_aspect('equal')

# visualize the level set
CS = plt.contour(A, B, levels, levels=np.logspace(0, 1, num=20), cmap=plt.cm.rainbow_r)
CB = plt.colorbar(CS, shrink=0.8, extend='both')

# visualize trajectory
plt.scatter(w_true[0], w_true[1], c='r')
plt.scatter(w_list[:, 0], w_list[:, 1])
plt.plot(w_list[:, 0], w_list[:, 1])
```

#### 4.1 Напишите функцию, которая вычисляет ошибку RMSE (root mean square error):
$$ RMSE(a,\mathbb X)=\sqrt{\frac{\sum_{i=1}^{N}(a(x_i)-y_i)^2}{N}}.$$


```python
def RMSE(a,X,Y):
  return np.sqrt((np.sum(a(X)-Y)**2)/X.shape[0])
#ваш код здесь
```

### 5. Отбор признаков в Lasso регрессии. 
Рассмотрим Lasso регрессию с различным параметром регуляризации $\alpha \in [10^{-3},10^3]$. Посмотрите, какое количество коэффициентов зануляется в линейной модели. Воспользуйтесь кроссвалидацией (например, ShuffleSplit), чтобы построить несколько алгоритмов для одного и того же значения  $\alpha$, а затем взять среднее число ненулевых коэффициентов по этим алгоритмам.  Постройте график зависимости (логарифмический масштаб по оси абсцис) среднего числа ненулевых коэффициентов от коэффицента регуляризации $\alpha$.

P.S. Обратите внимание на необходимость стандартизации при каждом разбиении выборки.  

```python
alphas = np.logspace(-3, 3,20)
epsilon = 1e-10
non_null = []

for a in alphas:

  lasso_model = Lasso(alpha = a)
  cv = ShuffleSplit(n_splits=7, test_size=0.3, random_state=42)
  non_null_weights = 0.0

  

  for train, test in cv.split(X):
    lscaler = StandardScaler()
    scaler.fit(X[train])
    x_train = scaler.transform(X[train])
    x_test = scaler.transform(X[test])
    lasso_model.fit(x_train, y[train])

    non_null_weights+=np.count_nonzero(np.abs(lasso_model.coef_) > epsilon)/7.0
    #print("Train score:",lasso_model.score(X_train,y[train]),"Test score:",lasso_model.score(X_test,y[test]))
  non_null.append(non_null_weights)

fig = plt.figure(figsize=(10,10))

plt.plot(alphas,non_null)
plt.title("Dependence of number of non_null weights on ln(alpha)", fontsize=15)
plt.xlabel("Natural logarithm of alpha",fontsize=15)
plt.ylabel("Number of non_null weights", fontsize=15)
plt.xscale("log")
plt.show()

#ваш код здесь
```
