###  Задача 1. (2 балла)
Позволяет ли предсказывать корректные вероятности:

1) квадратичная функция потерь $L(y, z) =(y-z)^2$, 

2) экспоненциальная функция потерь $L(y, z) = \exp(-yz)$ ?

  Рассмотрим сначала квадратичную функцию потерь:

$$
  L(y,z) = (y-z)^2
$$

  Предположим, что у нас имеется выборка $X_l$ размера $l$, а в качестве меток для классификации будем использовать $y=\{0,1\}$ вместо $y=\{-1,1\}$. Тогда функцию потерь на $i$-ом объекте можно переписать следующим образом.

$$
  L(y_i,z_i) = [y_i=1] \cdot (1-z_i)^2+[y_i=0] \cdot (z_i)^2
$$

  Рассмотрим матожидние функции потерь в точке x:

$$
  \mathbf{E}[L(y,z) |x] = \mathbf{E}[ [y=1] \cdot (1-z)^2+[y=0] \cdot (z)^2 | x] = p( y = 1 |x) \cdot (1-z)^2 +(1-p(y=1|x)) \cdot z^2 
$$

$$
  \mathbf{E}[L(y,z) |x] = p( y = 1 |x) -2 \cdot p \cdot ( y = 1 |x) \cdot z +z^2
$$

  Продифференцируем по z для максимизации правдоподобия:

$$
  \dfrac{\partial}{\partial z} ( p( y = 1 |x) -2 \cdot p \cdot ( y = 1 |x) \cdot z +z^2) = 0
$$

$$
  p( y = 1 |x)  = z 
$$

  Следовательно квадратичная функция потерь позволяет предсказывать корректные вероятности

  Теперь рассматорим экспоненциальную функцию потерь, но вернемся к $y=\{-1,1\}$

$$
  L(y,z) = \exp^{-yz}
$$

  Перепишем её следующим образом:

$$
  L(y,z) = [y=1] \cdot exp^{-z}+[y=-1] \cdot exp^{z}
$$

  Рассмотрим матожидание функции в точке х:

$$
  \mathbf{E}[L(y,z) |x]  = \mathbf{E}[[y=1] \cdot exp^{-z}+[y=-1] \cdot exp^{z} |x] = p(y=1|x) \cdot exp^{-z} +(1-p(y=1|x)) \cdot exp^{z} 
$$

  Продифференцируем по z:

$$
  \frac{\partial}{\partial z} (p(y=1|x) \cdot exp^{-z} +(1-p(y=1|x)) \cdot exp^{z})\\
  p(y=1|x) \cdot (exp^{z} +exp^{-z})=exp^z\\
  p(y=1|x) = 1-\frac{1}{exp^{2z}+1}\\
  exp^{2z}=\frac{1}{1-p}-1\\
  z=\frac{1}{2}ln(\frac{p}{1-p})
$$

  Значение предсказываемое алгоритмом в точке не совпадает с вероятностью принадлежности данной точки к классу $y=1$. Следовательно экспоненциальная функция потерь не позволяет предсказать корректные вероятности.

  Если же мы попробуем изменить метку класса с -1 на 0, то получим:

  $$ E[L(y,z)|x] = p(y=+1|x)\exp(-z) + (1-p(y=+1|x))\exp(-0*z)$$

Продифференцировав по z получаем: 

$$\frac{\partial}{\partial z}E[L(y,z)|x] = -z*p(y=+1|x)\exp(-z) = 0$$

Тогда алгоритм будет предсказание равное нулю, что не является корректным предсказанием вероятности.

### Задача 2.  (2 балла)
Рассмотрим два вектора, $v_1=(1,1,2)$ и $v_2=(1,1,1)$. Переход в спрямляющее пространство осуществляется с помощью гауссовской ядерной функции $K(v,v {'})=\exp(-\gamma|v-v{'}|^2)$.
*  Как от $\gamma$ зависят длины векторов в спрямляющем пространстве?
*  Как от $\gamma$ зависит угол между векторами в спрямляющем пространстве?

####Зависимость длины вектора в спрямляющем пространстве от $\gamma$
\\

Необходимо вычислить длину вектора $v$ в спрямляющем пространстве, если известно только ядро $K(v,v')$. Длина вектора в спрямляющем пространстве это его норма в этом пространстве, то есть необходимо найти норму вектора $\phi (v)$.

$$
  ||\phi(v)||=\sqrt{||\phi(v)||^2}=\sqrt{<\phi(v),\phi(v)>}=\sqrt{K(v,v)}
$$

Поэтому

$$
  ||\phi(v)|| = \sqrt{exp^{-j|v-v|^2}}=\sqrt{exp^{-j\cdot 0}}=\sqrt{1}=1
$$

Следовательно, длина любого вектора в спрямляющем пространстве равна 1, то есть она не зависит от $\gamma$. Длины обоих векторов, которые представлены выше равны 1 в спрямляющем пространстве.

\\

####Зависимость угла между векторами в спрямляющем пространстве от $\gamma$

Обозначим угол между векторами в спрямляющем пространстве через $\theta$

$$
  cos(\theta) = \frac{<\phi(v),\phi(v')>}{||\phi(v)|| \cdot ||\phi(v')||}\\
  cos(\theta) = \frac{K(v,v')}{\sqrt{K(v,v) \cdot K(v',v')}}\\
  cos(\theta) = \frac{exp^{-\gamma \cdot ||v-v'||^2}}{\sqrt{1 \cdot 1}}\\
  \theta = arccos(exp^{-\gamma \cdot ||v-v'||^2})
$$

При этом стоит заметить, что в силу неравенства треугольника для нормы получаем, что:

$$
  ||v-v'||<=||v||+||v'||=2\\
  0 \leq ||v-v'|| \leq 2
$$

Также косинус принадлежит интервалу от -1 до 1, то так как exp принимает только неотрицательные значения, то угол между векторами будет от 0 до 90 градусов, при этом, если $\gamma$>0 и большим, при норме не равной нулю, то угол будет расти до отметки в 90 градусов и соответственно при уменьшении $\gamma$ к нулю угол между векторами будет уменьшаться в сторону нуля градусов, при этом при отрицательном значении $\gamma$ выдаваемые значения будут некорректными, так как значение арккосинуса будет комплексным.


### Задача 3. (2 балла) 

Является ли функция $K(x,z)=\cos(x-z)$ ядром?

Рассмотрим косинус на возможность взятия его в качестве ядра.

$$
  cos(x-z) = cos(x) \cdot cos(z)+sin(z) \cdot sin(x)\\
  f_1(x) = cos(x), \;f_2(x)=sin(x)
$$

Известно, что одним из способов задания ядра является:

$$
  K(x,y) = f(x) \cdot f(y)
$$

Где $f(x): X \rightarrow R$, вещественнозначная функция, тогда

$$
  cos(x) \cdot cos(y) = f_1(x) \cdot f_1(x)\\
  K_1(x,y) = f_1(x) \cdot f_1(y)\\
$$
$$
  sin(x) \cdot sin(y) = f_2(x) \cdot f_2(y)\\
  K_2(x,y) = f_2(x) \cdot f_2(y)\\
$$
$$
  cos(x-z) = K_1(x,z)+K_2(x,z)\\
$$
Известно, что если $K_1(x,y)$ и $K_2(x,y)$ - ядра, заданные на множестве $X$, тогда функция $K(x,y)$, заданная как:

$$
  K(x,y) = K_1(x,y)+K_2(x,y) 
$$

Является ядром. Следовательно функция $K(x,z)=cos(x-z)$ является ядром.

### Задача 4. (4 балла)

```python
%pylab inline
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

Сгенерируем синтетические данные:

```python
X, y = make_classification(
    n_samples=10000, n_features=10, n_informative=5, n_redundant=5,
    random_state=42)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

####  1. Обучите метод опорных векторов. Воспользуйтесь классом [LinearSVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html).

```python
from sklearn.svm import LinearSVC

SVM_classifier = LinearSVC()# changing C here
SVM_classifier.fit(X_train, y_train)

pred = SVM_classifier.predict(X_test)
```

#### 2. Чему равны доля верных предсказаний, точность, полнота и F-мера классификатора? 

```python
from sklearn import metrics
print("Accuracy",metrics.accuracy_score(y_test,pred))
print("Recall",metrics.recall_score(y_test,pred))
print("Precision",metrics.precision_score(y_test,pred))
print("F1-score",metrics.f1_score(y_test,pred))
```

#### 3.  Постройте ROC кривую. Определите AUC-ROC. 
Обратите внимание, что метод опорных векторов  не умеет predict_proba, но имеет метод decision_function.
Определите вероятности принадлежности к положительному классу как:
$$P(y=+1|x)=\frac{a(x)-\min(a_i)}{\max(a_i)-\min(a_i)}$$

```python
dist = SVM_classifier.decision_function(X_test)
probs = (dist-min(dist))/(max(dist)-min(dist))

fpr, tpr, tresh = metrics.roc_curve(y_test,probs)


fig = plt.figure(figsize=(10,10))
plt.plot(fpr,tpr)
plt.xlabel("False positive rate",fontsize=15)
plt.ylabel("True positive rate",fontsize=15)
plt.title("Roc-curve",fontsize=15)
plt.show()

print("AUC-ROC", metrics.roc_auc_score(y_test,probs))
```

#### 4. Калибровка вероятностей.  

Перейдём к оценке качества выдаваемых алгоритмами вероятностей. Начнём с калибровочных кривых.

Допустим, алгоритм возвращает некоторые числа от нуля до единицы. Хорошо ли они оценивают вероятность? Для этого разобьем отрезок $[0, 1]$ на несколько маленьких отрезков одинаковой длины. Рассмотрим $i$-й отрезок с границами $[a_i, b_i]$ и предсказания $p_1, p_2, \dots, p_k$, которые попали в него. Пусть им соответствуют истинные ответы $y_1, y_2, \dots, y_k$. Если алгоритм выдает корректные вероятности, то среди этих истинных ответов должно быть примерно $(a_i + b_i) / 2$ единиц. Иными словами, если нарисовать кривую, у которой по оси X отложены центры отрезков, а по оси Y — доли единичных ответов этих в отрезках, то она должна оказаться диагональной. 
#### Нарисуйте для найденных вероятностей калибровочную кривую, сравните с диагональю. 
Воспользуйтесь [calibration_curve](https://scikit-learn.org/stable/modules/generated/sklearn.calibration.calibration_curve.html).

###Без калибровочной кривой

```python
def diag(data, n):
  x = np.arange(0,1+1/n,1/n)
  eyes = []
  for i in range(1, len(x)):
    a_i = x[i-1]
    b_i = x[i]
    tmp_1 = data[(data>=a_i) & (data<=b_i) & (y_test==1)]
    tmp_2 = data[(data>=a_i) & (data<=b_i)]
    eyes.append(np.count_nonzero(tmp_1)/np.count_nonzero(tmp_2))
  _ = plt.figure(figsize=(10,10))
  plt.plot((x[1:]+x[:-1])/2,eyes)
  plt.xlabel("Centers of intervals",fontsize=15)
  plt.ylabel("Number of ones in the interval", fontsize =15)
  plt.show()


diag(probs,20)
```

### С калибровочной кривой

```python
from sklearn.calibration import CalibratedClassifierCV, calibration_curve

fraction, mean = calibration_curve(y_test, probs, n_bins=20)
plt.figure(figsize=(10,10))

plt.plot(mean, fraction, "s-", label="SVC")

plt.plot([0, 1], [0, 1], "k:", label="Calibrated")
plt.ylabel("Пропорция положительного класса")
plt.xlabel("Среднее предсказанное значение")
plt.title("График откалиброванной кривой")
axes = plt.gca()
axes.set_ylim([-0.1, 1.1])
plt.legend()
plt.show()
```


Воспользуйтесь [CalibratedClassifierCV](https://scikit-learn.org/stable/modules/generated/sklearn.calibration.CalibratedClassifierCV.html#sklearn.calibration.CalibratedClassifierCV)из sklearn для калибровки вероятностей метода опорных векторов на обучении и постройте с его помощью предсказания для тестовой выборки. Нарисуйте для них калибровочную кривую. Улучшилась ли она?

```python
model = LinearSVC()
calibrated = CalibratedClassifierCV(model, method='sigmoid', cv=5)
calibrated.fit(X_train, y_train)

probs = calibrated.predict_proba(X_test)[:, 1]
fop, mpv = calibration_curve(y_test, probs, n_bins=15, normalize=True)

_ = plt.figure(figsize=(10,10))
plt.plot([0, 1], [0, 1], linestyle='--',label="Calibrated")
plt.plot(mpv, fop, marker='.',label="SVC")
plt.title("Calibrated curve", fontsize=15)
plt.legend()
plt.show()
```

Построенная таким образом калибровочная кривая располагается заметно ближе к диагональной линии, что свидетельствует об улучшении результата.
