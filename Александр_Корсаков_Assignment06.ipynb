### 1. Энтропия (3 балла)
Вычислите энтропию и индекс Джини случайной величины $X$, если 

* величина $X$ равновероятно принимает $k\ge2$  значений;

* величина $X$ равномерно распределена на отрезке $[0; a]$;

* величина $X$ нормально распределена $\mathcal N (0, \sigma^2)$;

### Формулы:

#### 1)Индекс Джини:

Для дискретного числа значений:
$$
F_G(R_m)=\sum_{k = 1}^{K} p_{k} (1 - p_{k}).
$$

Для непрерывной случайной величины:
$$
F_G(R_m)=\int_{-\inf}^{+\inf} f(x)(1 - f(x))dx.
$$

#### 2) Энтропия.
Для дискретного числа значений:
$$
F_H(R_m)=-\sum_{k = 1}^{K} p_{k} \log_2 p_{k}.
$$

Для непрерывной случайной величины:
$$
F_H(R_m)=-\int_{-\inf}^{+\inf} f(x) \log_2 f(x)dx.
$$

\\

###1. Величина  X  равновероятно принимает  k≥2  значений

####Индекс Джини:

$$
F_G(R_m)=\sum_{k = 1}^{K} p_{k} (1 - p_{k})=\sum_{k = 1}^{K} \frac{1}{K} (1 - \frac{1}{K})=\frac{K-1}{K}
$$

\\

####Энтропия.
$$
F_H(R_m)=-\sum_{k = 1}^{K} p_{k} \log_2 p_{k}=-\sum_{k = 1}^{K} \frac{1}{K} \log_2 \frac{1}{K}=-\log_2 \frac{1}{K}
$$

\\

###2. Величина X равномерно распределена на отрезке  [0;a]

####Индекс Джини:

$$
F_G(R_m)=\int_{-\inf}^{+\inf} f(x)(1 - f(x))dx=\int_{0}^{|a|}\frac{1}{|a|}(1-\frac{1}{|a|})dx=\frac{|a|-1}{|a|}
$$

####Энтропия:
$$
F_H(R_m)=-\int_{-\inf}^{+\inf} f(x) \log_2 f(x) = -\int_{0}^{|a|}\frac{1}{|a|}\log_2\frac{1}{|a|}dx=-\log_2\frac{1}{|a|}
$$

Вероятно предполагается, что a больше или равно 1

\\

###  2. Критерии информативности (2 балла)
На лекциях говорилось, что критерий информативности для набора объ-
ектов R вычисляется на основе того, насколько хорошо их целевые переменные предсказываются константой (при оптимальном выборе этой константы):
$$
H(R)=\frac{1}{|R|}\sum_{(x_i,y_i)\in R} L(y_i,c),
$$
где $L(y, c)$ - некоторая функция потерь и
$$
c=argmin_cH(R).
$$
 
Соответственно, чтобы получить вид критерия при конкретной функции потерь, необходимо аналитически найти оптимальное
значение константы и подставить его в формулу для H(R).
Выведите критерии информативности для следующих функций потерь:

* $L(y,c)=\sum_{k=1}^K (c_k-[y=k])^2$,

* $L(y,c)=-\sum_{k=1}^K [y=k] \log c_k$. 

Для квадратичной функции потерь
$$
  H(R)=\frac{1}{|R|}\cdot \sum_{(x_i,y_i) \in R} \sum_{k=1}^K (c_k-[y_i=k])^2\\
  H(R) \rightarrow min\\
  \frac{\partial{H}}{\partial c_k}=0\\
  \frac{\partial{H}}{\partial c_k}=\frac{1}{|R|}\cdot \sum_{(x_i,y_i) \in R} 2(c_k-[y_i=k])=0\\
  c_k=\frac{1}{|R|}\cdot\sum_{(x_i,y_i) \in R} [y_i=k]\\
  c_k=p_k
$$
Подставив получаем
$$
  H(R)=\sum_{k=1}^K p_k(1-p_k)
$$

\\

$$
   H(R)=\frac{1}{|R|}\cdot \sum_{(x_i,y_i) \in R} \sum_{k=1}^K [y_i=k] \log c_k\\
   H(R) \rightarrow min
$$

\\

Задача на поиск экстремума с ограничением $\sum_{k=1}^K c_k=1$. Поэтому надо искать минимум Лагранжиана.

$$
  L(c,\lambda) = -\frac{1}{|R|}\cdot\sum_{(x_i,y_i) \in R} \sum_{k=1}^K [y_i=k] \log c_k+\lambda\sum_{k=1}^K c_k \rightarrow min_{c_k}\\
  \frac{\partial L(c,\lambda)}{\partial c_k} = -\frac{1}{|R|}\sum_{(x_i,y_i) \in R} \frac{[y_i=k]}{c_k}+\lambda=-\frac{p_k}{c_k}+\lambda=0\\
  c_k=\frac{p_k}{\lambda}; \;\sum_{k=1}^K c_k=1 \rightarrow \lambda=1\\
  c_k=p_k
$$
Подставив получаем
$$
  H(R)=-\sum_{k=1}^K p_k\log(p_k)
$$



### 3.  Переобучение и неустойчивость решающих деревьев (3 балла)
* сгенерируйте случайную выборку $X=\{x_1,x_2\}$ с количеством объектов $N=100$: $x_1,x_2$ из нормального распределения, целевая переменная $y=\sqrt{x_1^2+x_2^2}$;

* изобразите выборку на плоскости, по осям - значения признаков $x_1$, $x_2$, цвет отражает значение целевой переменной;

* постройте решающее дерево для задачи регрессии с параметрами по умолчанию;

* с помощью get_grid постройте разделяющую границу регрессора;

* постройте деревья-регрессоры  для разных значений глубины дерева и минимального числа объектов в листе;

* изобразите на одном графике  соответствующие разделяющие границы регрессоров (например, с помощью plt.subplot2grid);

* Решающие деревья — это алгоритмы, неустойчивые к изменениям обучающей выборки, т.е. при малейших её изменениях итоговый классификатор может радикально измениться. Посмотрите, как будет меняться структура дерева при обучении на разных 90%-х подвыборках (постройте три регрессора и соответствующие разделяющие границы регрессоров)

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor

def get_grid(data): 
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), # создаем координатную матрицу c шагом 0.01
                         np.arange(y_min, y_max, 0.01))

jet=plt.get_cmap('coolwarm')

X = np.random.normal(size=(100,2))
Y = np.sqrt((X**2).sum(axis=1))


plt.scatter(X[:,0], X[:,1], s=100, c=Y, cmap=jet)
plt.title("Sample illustration", fontsize=15)
plt.xlabel("X1", fontsize=13)
plt.ylabel("X2", fontsize=13)
plt.colorbar()
plt.show()


tree_model = DecisionTreeRegressor()
tree_model.fit(X, Y)

xx, yy = get_grid(X)

predicted = tree_model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 8))
plt.pcolormesh(xx, yy, predicted, cmap='spring')
plt.scatter(X[:,0], X[:,1], c=Y, cmap='jet', alpha = 0.9)
plt.title("Hyperplanes chart (version 1)", fontsize=15)
plt.colorbar()
plt.xlabel("x1", fontsize=15)
plt.ylabel("x2", fontsize=15)
plt.show()

fig = plt.figure(figsize=(8, 8))
plt.pcolormesh(xx, yy, predicted, cmap=plt.cm.YlOrRd)
plt.scatter(X[:,0], X[:,1],  c='g')
plt.title("Hyperplanes chart (version 2)", fontsize=15)
plt.xlabel("x1", fontsize=15)
plt.ylabel("x2", fontsize=15)
plt.show()
```

```python
max_depth=[2, 4, None]
min_samples_leaf=[1,2,15]

regressors = []
for depth in max_depth:
  for leaf in min_samples_leaf:
    dt = DecisionTreeRegressor(max_depth = depth,min_samples_leaf=leaf)
    dt.fit(X, Y)
    predicted = dt.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    regressors.append(predicted)


fig, axs = plt.subplots(3, 3 )
fig.set_size_inches((20,20))

for i in range(3):
  for j in range(3):
    axs[i,j].pcolormesh(xx, yy, regressors[3*i+j], cmap='spring')
    axs[i,j].set_title("Samp per leaf "+str(min_samples_leaf[j])+' '+"depth "+str(max_depth[i]))


for ax in axs.flat:
    ax.set(xlabel='X1', ylabel='X2')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
from sklearn.model_selection import train_test_split


fig, axs = plt.subplots(3, 1)
fig.set_size_inches((6,18))

for i in range(3):
  X_train, _, y_train, _ = train_test_split(X, Y, test_size=0.1, random_state=np.random.randint(1,1000))

  dt = DecisionTreeRegressor(max_depth = 4)
  dt.fit(X_train, y_train)
  xx, yy = get_grid(X_train)
  predicted = dt.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

  axs[i].pcolormesh(xx, yy, predicted, cmap='spring')

for ax in axs.flat:
    ax.set(xlabel='X1', ylabel='X2')
```

### 4. Решающее дерево для регрессии на данных (2 балла)
* разделите выборку на обучающую и тестовую в соотношении $7:3$;
* с помощью GridSearch подберите наилучшие параметры для решающего дерева регрессора (выведите параметры наилучшего регрессора). 

```python
from sklearn.datasets import load_boston
data = load_boston()

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
scoring='r2'

max_features=[None, 'log2', 'sqrt']

max_depth =[2, 4, 6, 8, 10, 20, 50]

min_samples_leaf=np.arange(1,20)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn import metrics

x = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.7, random_state =1)


param_grid=[{'decisiontreeregressor__max_depth':max_depth,
             'decisiontreeregressor__max_features':max_features,
              'decisiontreeregressor__min_samples_leaf':min_samples_leaf}]

pipe_tree = make_pipeline(DecisionTreeRegressor(random_state=1))
gs = GridSearchCV(estimator=pipe_tree, param_grid=param_grid, scoring=scoring, cv=10)

gs = gs.fit(X_train, y_train)
```

```python
print(gs.best_params_)
model = gs.best_estimator_
predicted = model.predict(X_test)
metrics.r2_score(y_test, predicted)
```
